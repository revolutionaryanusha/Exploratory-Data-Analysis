# New features for Date column

```python
# StartTime(UTC)	EndTime(UTC)
df["StartTime(UTC)"]=pd.to_datetime(df["StartTime(UTC)"])
df['Start_date']=df['StartTime(UTC)'].dt.date
df['Start_time']=df['StartTime(UTC)'].dt.time
df['Start_year']=df['StartTime(UTC)'].dt.year
df['Start_month']=df['StartTime(UTC)'].dt.month
df['Start_hour']=df['StartTime(UTC)'].dt.hour
df['Start_weekend']=df['StartTime(UTC)'].dt.weekday
df['Start_weekend']=df['StartTime(UTC)'].dt.day_name()
df['Start_month']=df['StartTime(UTC)'].dt.month_name()
df['SWeek'] = df['StartTime(UTC)'].dt.isocalendar().week

df["EndTime(UTC)"]=pd.to_datetime(df["EndTime(UTC)"])
df['End_date']=df['EndTime(UTC)'].dt.date
df['End_time']=df['EndTime(UTC)'].dt.time
df['End_year']=df['EndTime(UTC)'].dt.year
df['End_month']=df['EndTime(UTC)'].dt.month
df['End_hour']=df['EndTime(UTC)'].dt.hour
df['End_weekend']=df['EndTime(UTC)'].dt.weekday
df['End_weekend']=df['EndTime(UTC)'].dt.day_name()
df['End_month']=df['EndTime(UTC)'].dt.month_name()
df['EWeek'] = df['EndTime(UTC)'].dt.isocalendar().week
df['Duration'] = df.apply(lambda row: (datetime.datetime.combine(datetime.date.today(), row['End_time']) - 
                                      datetime.datetime.combine(datetime.date.today(), row['Start_time'])), axis=1)

```

# C. You must perform PCA for N-Component 3. You must Encode Features to Numerical representation. 10 Marks

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoded_df = df.apply(encoder.fit_transform)
encoded_df
```

```python

# Perform PCA with 3 components
pca = PCA(n_components=3)
pca_result = pca.fit_transform(encoded_df)

# Store the PCA results in a new dataframe
df_pca = pd.DataFrame(data=pca_result, columns=['PCA_1', 'PCA_2', 'PCA_3'])
df_pca["y"]=df["Precipitation(in)"]
df_pca
```

```python
import matplotlib.pyplot as plt

x=df_pca['PCA_1']
y=df_pca["y"]

plt.scatter(x, y)
plt.xlabel('prediction ')
plt.ylabel('PCA1')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=df_pca['PCA_2']
y=df_pca["y"]

plt.scatter(x, y)
plt.xlabel('prediction ')
plt.ylabel('PCA1')
plt.show()
```
